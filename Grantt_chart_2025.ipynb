```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Tasks and their durations
tasks = [
    "Complete 3 international AI skin papers",
    "Submit 2 ISI papers for Nafosted",
    "Submit AI skin patent",
    "Midterm review of AI skin project",
    "Final review of Nafosted project",
    "Develop Onelab NGS-qPCR platform for STI and lung cancer",
    "Complete bioinformatics training platform",
    "Save $24,000 from Japan AI project",
    "Save 200 million VND for personal safety fund",
    "Establish Can Tho Song Nuoc Company (3000 cosmetic samples)",
    "Achieve IELTS score of 7.5",
    "Teach 200 hours in 2025"
]

start_dates = [
    datetime(2025, 1, 1), datetime(2025, 1, 1), datetime(2025, 1, 1), datetime(2025, 1, 1),
    datetime(2025, 7, 1), datetime(2025, 2, 1), datetime(2025, 4, 1), datetime(2025, 1, 1),
    datetime(2025, 1, 1), datetime(2025, 1, 1), datetime(2025, 7, 1), datetime(2025, 1, 1)
]

end_dates = [
    datetime(2025, 12, 31), datetime(2025, 10, 31), datetime(2025, 8, 31), datetime(2025, 6, 30),
    datetime(2025, 12, 31), datetime(2025, 5, 31), datetime(2025, 7, 31), datetime(2025, 12, 31),
    datetime(2025, 12, 31), datetime(2025, 12, 31), datetime(2025, 11, 30), datetime(2025, 12, 31)
]

# Assigning different colors to represent key tasks
colors = [
    "steelblue",  # Complete 3 international AI skin papers
    "mediumseagreen",  # Submit 2 ISI papers for Nafosted
    "gold",  # Submit AI skin patent
    "lightcoral",  # Midterm review of AI skin project
    "darkorange",  # Final review of Nafosted project
    "violet",  # Develop Onelab NGS-qPCR platform for STI and lung cancer
    "lightblue",  # Complete bioinformatics training platform
    "slategray",  # Save $24,000 from Japan AI project
    "limegreen",  # Save 200 million VND for personal safety fund
    "palegoldenrod",  # Establish Can Tho Song Nuoc Company (3000 cosmetic samples)
    "dodgerblue",  # Achieve IELTS score of 7.5
    "plum"  # Teach 200 hours in 2025
]

# Assigning tasks to phases to reduce overlap
phases = {
    "Phase 1": [0, 1, 7, 8],
    "Phase 2": [2, 5, 6],
    "Phase 3": [3, 9],
    "Phase 4": [4, 10, 11]
}

# Creating the Gantt chart
fig, ax = plt.subplots(figsize=(14, 10))

for phase, task_indices in phases.items():
    for i in task_indices:
        ax.barh(f"{phase} - {tasks[i]}", (end_dates[i] - start_dates[i]).days, left=start_dates[i], color=colors[i])

# Formatting the chart
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.xticks(fontsize=12, rotation=45)  # Adjust font size and rotation for x-axis
plt.yticks(fontsize=12)  # Adjust font size for y-axis
plt.xlabel("Timeline", fontsize=14)
plt.ylabel("Tasks by Phases", fontsize=14)
plt.title("2025 Gantt Chart for Project Timeline (Grouped by Phases)", fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the chart
plt.show()
```
